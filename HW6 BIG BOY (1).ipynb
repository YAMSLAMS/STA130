Simple Linear Regression is a way to explain the relationship between two variables, the predictor variable x_i and the outcome variable denoted Y_i. The predictor variable is an independent variable that is believed to influence the outcome. The outcome variable is dependent on the predictor variable and is the thing that is trying to be predicted. The slope B_1 is the amount that Y_i changes on average per one-unit increase of X. B_0 is the value of Y_i when x_i = 0. It represents the starting point of the regression line. The error term Îµ accounts for the difference between the observed value of 
Y_i and the value predicted by the model.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats

# Set seed for reproducibility
np.random.seed(42)

# Define the parameters of the simple linear regression model
beta_0 = 5  # Intercept
beta_1 = 2  # Slope
sigma = 1.5  # Standard deviation of the errors (sqrt of variance)

# Generate predictor variable X (independent variable)
X = np.linspace(0, 10, 100)  # 100 points between 0 and 10

# Generate random normal errors with mean 0 and standard deviation sigma
errors = np.random.normal(0, sigma, size=X.shape)

# Calculate outcome variable Y (dependent variable) based on the regression model
Y = beta_0 + beta_1 * X + errors

# Plot the generated data and the underlying true regression line
plt.figure(figsize=(8, 6))

# Plot the observed Y values (with errors)
plt.scatter(X, Y, label='Observed Data (Y)', color='blue')

# Plot the true regression line (without errors)
Y_true = beta_0 + beta_1 * X  # True line without noise
plt.plot(X, Y_true, color='red', label=r'True Line $Y = \beta_0 + \beta_1 X$', linewidth=2)

# Add labels and title
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.title('Simple Linear Regression: True Line vs Observed Data')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python
2.

import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Set seed for reproducibility
np.random.seed(42)

# Define parameters for the simple linear regression model
beta_0 = 5   # Intercept
beta_1 = 2   # Slope
sigma = 1.5  # Standard deviation of errors

# Generate predictor variable X
X = np.linspace(0, 10, 100)

# Generate random normal errors
errors = np.random.normal(0, sigma, size=X.shape)

# Calculate outcome variable Y
Y = beta_0 + beta_1 * X + errors

# Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Print the model summary
print(model.summary())

# Create predictions for visualization
data['Y_pred'] = model.predict(data['X'])

# Plot the observed data and the fitted regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue', alpha=0.5)
plt.plot(data['X'], data['Y_pred'], color='red', label='Fitted Line', linewidth=2)
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.title('Fitted Simple Linear Regression Model')
plt.legend()
plt.show()
```

The difference between the fitted and true lines lies in how we got both of them. The true line is based on the true parameter used to simulate the data. Since this line is based on the parameters B_1 and B_0, it does not account for the random error term and remains fixed always. In contrast, the fitted line is not fixed. The fitted line is derived from thhe dataset whicch includes noise thus it can only estimatte the rellationship between the X_i and Y_i variables. The more noise a dataset has, the further the fitted line will be from the true line, so as the size of a dataset increases, the noise decreases as it averages out over more points, and in turn, the fitted line gets closer to the true line.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import statsmodels.api as sm
import statsmodels.formula.api as smf

# Set seed for reproducibility
np.random.seed(42)

# Define parameters for the simple linear regression model
beta_0 = 5   # Intercept
beta_1 = 2   # Slope
sigma = 1.5  # Standard deviation of errors

# Generate predictor variable X
X = np.linspace(0, 10, 100)

# Generate random normal errors
errors = np.random.normal(0, sigma, size=X.shape)

# Calculate outcome variable Y
Y = beta_0 + beta_1 * X + errors

# Create a pandas DataFrame
data = pd.DataFrame({'X': X, 'Y': Y})

# Fit the Simple Linear Regression model using statsmodels
model = smf.ols(formula='Y ~ X', data=data).fit()

# Create predictions for visualization
data['Y_pred'] = model.predict(data['X'])

# Calculate the true regression line without noise
Y_true = beta_0 + beta_1 * X

# Plot the observed data, fitted regression line, and true regression line
plt.figure(figsize=(10, 6))
plt.scatter(data['X'], data['Y'], label='Observed Data', color='blue', alpha=0.5)
plt.plot(data['X'], data['Y_pred'], color='red', label='Fitted Line', linewidth=2)
plt.plot(data['X'], Y_true, color='green', label='True Line', linestyle='--', linewidth=2)
plt.xlabel('Predictor (X)')
plt.ylabel('Outcome (Y)')
plt.title('Fitted Simple Linear Regression Model with True Line')
plt.legend()
plt.show()
```

Null Hypothesis: There is no linear association between waiting time and eruption duration. Mathematically, this can be stated as B_1 = 0, as a slope of 0 implies that there is no change in duration with respect to a change in waiting time.

Alternate Hypothesis: There is a linear association between waiting time and eruption duration. Mathematically, this can be stated as B_1 does not equal 0, as any slope that is not 0 implies some change in the duration of the waiting time changes.

In this analysis, we will use a two-tailed test for the intervals of 62, 64, and 66 minutes. We will fit a regression line to each wait time and calculate the p-values. Afterwards, we will compare the p-values with predetermined alpha values (commonly set at 0.05) to determine whether to reject the null hypothesis or fail to reject it in favour of the alternative hypothesis.

```python
import seaborn as sns
import statsmodels.formula.api as smf
import plotly.express as px

# Load dataset
old_faithful = sns.load_dataset('geyser')

# Analyze for different short wait limits
short_wait_limits = [62, 64, 66]
results = {}

for limit in short_wait_limits:
    short_wait = old_faithful.waiting < limit
    model = smf.ols('duration ~ waiting', data=old_faithful[short_wait])
    fitted_model = model.fit()
    summary = fitted_model.summary().tables[1]
    
    # Create scatter plot with regression line
    fig = px.scatter(old_faithful[short_wait], x='waiting', y='duration', 
                     title=f"Old Faithful Geyser Eruptions for short wait times (<{limit})", 
                     trendline='ols')
    fig.show(renderer="png")
    
# Create a dictionary to store p-values
p_values = {}

for limit in short_wait_limits:
    # Filter the dataset for waiting time less than the current limit
    short_wait = old_faithful[old_faithful['waiting'] < limit]
    
    # Fit the linear regression model
    model = smf.ols('duration ~ waiting', data=short_wait)
    fitted_model = model.fit()
    
    # Extract the p-value for the waiting coefficient
    p_value = fitted_model.pvalues['waiting']
    p_values[limit] = p_value

# Display the p-values
for limit, p_value in p_values.items():
    print(f'Short Wait Limit: {limit} minutes, p-value: {p_value}')
```

For the wait times of 62, 64, and 68 minutes, the p-values were calculated to be 0.2384, 0.0362, and 0.0003, respectively. At a wait time of 62 minutes, the p-value shows week evidence to reject the null hypothesis. At a wait time of 64 minutes, the p-value shows moderate evidence to reject the null hypothesis. At a wait time of 66 minutes, the p-value shows strong evidence to reject the null hypothesis. 

So, we have one interval of time that indicates that the null hypothesis shouldn't be rejected and two that show evidence that it should. What we can extrapolate from this information is that at lower wait times, there is little to no evidence to reject the null hypothesis, but as the wait time increases, the evidence against the null hypothesis gets stronger. 

Thus, I conclude that at short wait times like 62 minutes, there is not sufficient evidence to reject the null hypothesis, but at wait times of 64 and 66 minutes, there is enough evidence to reject the null hypothesis.
